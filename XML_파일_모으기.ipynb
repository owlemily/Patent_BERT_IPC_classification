```python
import os
import shutil

# 구글 드라이브 경로 설정
base_dir = '/content/drive/MyDrive/특허데이터/20190121'  # "20190115" 폴더 경로
output_dir = '/content/drive/MyDrive/특허데이터/20190121_only_xml'  # XML 파일을 저장할 경로

# 출력 디렉터리가 없으면 생성
os.makedirs(output_dir, exist_ok=True)

# XML 파일만 검색하고 복사
for root, dirs, files in os.walk(base_dir):
    if "XML" in root:  # XML 폴더만 탐색
        for file in files:
            if file.endswith('.xml'):  # XML 파일만 필터링
                source_path = os.path.join(root, file)
                destination_path = os.path.join(output_dir, file)
                shutil.copy(source_path, destination_path)
                print(f"Copied: {source_path} -> {destination_path}")

print("모든 XML 파일이 성공적으로 복사되었습니다!")
```

## 파일 개수 확인

```python
import os
# 디렉토리 개수를 확인할 경로
target_dir = "/content/drive/MyDrive/특허데이터/20190121"
# 해당 디렉터리 내 디렉토리 개수 세기
dir_count = len([d for d in os.listdir(target_dir) if os.path.isdir(os.path.join(target_dir, d))])
print(f"'{target_dir}'에는 {dir_count}개의 디렉토리가 있습니다.")
```
